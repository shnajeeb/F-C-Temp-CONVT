```python
# Step 1: Install required libraries
!pip install pdf2image pytesseract
!apt-get install -y poppler-utils

# Step 2: Import necessary libraries
from google.colab import files
from pdf2image import convert_from_path
import pytesseract
import matplotlib.pyplot as plt
import os

# Step 3: Upload the PDF file
print("Please upload your PDF file:")
uploaded = files.upload()

# Extract the uploaded file name
file_name = list(uploaded.keys())[0]
print(f"Uploaded file: {file_name}")

# Step 4: Convert the PDF to images
images = convert_from_path(file_name)
print(f"Converted {len(images)} pages into images.")

# Step 5: Extract text from each page and display the images with the text
for page_num, image in enumerate(images, start=1):
    # Display the image
    plt.figure(figsize=(8, 10))
    plt.imshow(image)
    plt.axis('off')
    plt.title(f"Page {page_num}")
    plt.show()

    # Extract text using pytesseract
    text = pytesseract.image_to_string(image)
    print(f"Extracted text from Page {page_num}:\n")
    print(text)
    print("\n" + "="*80 + "\n")

# Step 6: Clean up (optional)
# Delete the uploaded file if needed to free up space
os.remove(file_name)
print("Process completed. Uploaded file has been removed.")
```
